1. Utilize um método grayscale e salve a imagem resultante no disco:

```python

import numpy as np
from PIL import Image

# Carregar a imagem frog.jpeg:
img_np = np.array(Image.open('frog.jpeg'))

# Calcular média dos canais RGB ao longo do eixo 2 (páginas):
img_avg2 = np.average(img_np[:, :, :3], axis=2).astype(np.uint8)

# Salvar a imagem resultante no disco:
gray_image = Image.fromarray(img_avg2)
gray_image.save('frog_gray.jpeg')
```

2. Converta uma imagem para binária, utilizando um método de grayscale diferente do anterior. Salve-a no disco:

```python
import numpy as np
from PIL import Image

# Funcao para converter imagem grayscale para binaria:
def grayscale_to_binary(gray_image, limiar):
    binary_image = np.zeros_like(gray_image)
    binary_image[gray_image > limiar] = 255  # Pixel na imagem em cinza > limiar = Pixel Branco
    
    return binary_image

# Carregar a imagem frog_gray.jpeg:
gray_image = np.array(Image.open('frog_gray.jpeg'))

valor_limiar = 127

# Converter a imagem em cinza para binaria:
binary_image = grayscale_to_binary(gray_image, valor_limiar)

# Salvar a imagem binaria no disco:
Image.fromarray(binary_image.astype(np.uint8)).save('frog_binary.jpeg')
```

3. Converta uma imagem para algum espectro monocromático utilizando um método de grayscale diferente dos anteriores. Exiba-a com o matplotlib.

      Dica: Caso necessário, uma cópia de uma imagem `img` pode ser feita com `img.copy()`.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Carregar a imagem
image = np.array(Image.open('frog.jpeg'))

# Converter para espectro monocromático
monochrome_image = np.mean(image, axis=2).astype(np.uint8)

# Exibir imagem monocromática
plt.imshow(monochrome_image, cmap='gray')
plt.axis('off')
plt.show()
```
